```python
import plotly.express as px
import pandas as pd

# reading the .csv file using read_csv() function from panda
real_estate = pd.read_csv("real-estate-sales_main.csv")

# grouping the data and calculating average by  using the  groupby)() and mean() function of panda
per_location_average_sale = real_estate.groupby(['StreetNameAndWay']).SalePrice.mean()

#wrting the data into .csv using the to_csv() of panda
per_location_average_sale.to_csv('average_sale_per_location.csv')

#reading a comma-separated values of .csv file  into dataframe 
data = pd.read_csv('average_sale_per_location.csv')

# constructing  dataframe out of csv data 
df = pd.DataFrame(data)

# renaming SalePrice of the dataframe to AverageSalePrice after averaging the sale 
new_df = df.rename(columns={'SalePrice':'AverageSalePrice'})

# creating lists out of dataframe column for the bar plot 

Locations = list(df.iloc[:, 0])
Sales = list(df.iloc[:, 1])

#using plotly for ploting bar plots
fig = px.bar(new_df,x=Locations, y=Sales,color = 'AverageSalePrice')
#Showing the plots
fig.show()

    
```
