```python
#import matplotlib for plotting plots
import matplotlib.pyplot as plt

#import panda for reading csv file and making a dataframe 
import pandas as pd 

#reading a comma-separated values into dataframe
data = pd.read_csv('2ndQt_2019.csv')

# constructing  dataframe out of csv data 
df = pd.DataFrame(data)

# preating lists out of dataframe column for the bar plot  
Dates = list(df.iloc[:, 0])
Sales = list(df.iloc[:, 1])

# setting the size of the plot
f = plt.figure()
f.set_figwidth(7)
f.set_figheight(9)

# assigning different colors for each bar
C = ["blue","orange","tab:olive","tab:brown"] 


# Plot the data using bar() method
plt.bar(Dates,Sales, width=0.8, color=C)

# assigning  values per plot
for i in range(len(Dates)):
    plt.text(i,Sales[i],Sales[i],ha="center",va="bottom")
    
#assigning title of the plot    
plt.title("2nd Quarter Sales Of Real Estate 2019 ")

#naming the axes
plt.xlabel("2nd Quarter Sale Dates")
plt.ylabel("Sum of Sale Price ")

# the below function to rotate and right allign the dates
plt.gcf().autofmt_xdate() 

# Show the plot
plt.show()
```
